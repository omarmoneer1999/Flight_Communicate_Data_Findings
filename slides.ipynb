```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe
df_8 = pd.read_csv('2008.csv')
```

## Delay Ratios

> In our first visualization we worked on showing the ratios of the each cause of the delay and how it 
will affect the flights.
### Flight cancelation or delay reasons and how much they contribute to the number of canceled flights

```python
#Here we will work on calculating the ratio of each delay reason.
# We will work on data preperation
tot_delays = df_8['WeatherDelay'].sum() + df_8['CarrierDelay'].sum()+df_8['SecurityDelay'].sum()+df_8['NASDelay'].sum()+df_8['LateAircraftDelay'].sum()
#We will work on the calculation 
wd = (df_8['WeatherDelay'].sum())/tot_delays
cd = (df_8['CarrierDelay'].sum())/tot_delays
sd = (df_8['SecurityDelay'].sum())/tot_delays
nd = (df_8['NASDelay'].sum())/tot_delays
ld = (df_8['LateAircraftDelay'].sum())/tot_delays
delay_ratios = [wd,cd,sd,nd,ld]

#Here we will work on making the visualization and the titles. 
lab = ['Weather Delays','Carrier Delays' , 'NAS Delays', 'Secuirity Delays','Late Air Craft']
plt.pie(delay_ratios,labels = lab, autopct='%.2f%%');
```

## Count Of Flights

> Here we will work on the count of the fights that released from each airlines. 
And it will shows what is the biggest share of each Carrier

```python
# We will work on making the titles and the visualization.
sb.countplot(y='UniqueCarrier',data = df_8, color ='black',order = df_8['UniqueCarrier'].value_counts().index);
plt.title('The number of the unique flights that made by the Carrier')
```

## Delays due to Weather

> This visualization will show us the delay amount and its minutes and its relation with the change due to the month. 

```python
#We will work on prepaaring the data to use it in the visualization
dfweat = df_8[['Month', 'WeatherDelay']].groupby('Month').sum().reset_index()
dfweat.plot(x='Month', y='WeatherDelay' )

#We will work on making the titles and the show of the visulaization.
plt.xticks(list(range(0,13,1)))
plt.ylabel("weather delays in minutes")
plt.title("Number of weather delays per month")
plt.show()
```
